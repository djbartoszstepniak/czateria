Zadanie 1.
***Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:***


*   Uważam, że wypełnianie PIT może zostać zautomatyzowane przez AI, z racji na to, że jest to praca na czystych danych.
*   Moim zdaniem streszczanie tekstu na chwilę obecną wymaga inteligencji człowieka, ponieważ AI bardzo często pomija ważne elementy.
* Tłumaczenie tekstu wymaga inteligencji człowieka, szczególnie kiedy jest on bardziej skomplikowany i abstrakcyjny.
* Klasyfikacja tekstu do kategorii tematycznych nie wymaga raczej inteligencji człowieka, jest to dość proste zadanie.
* Odpowiadanie na proste pytania nie wymaga inteligencji człowieka. AI radzi sobie z tym całkiem dobrze.
* Układanie rozkładu komunikacji miejskiej jest zadaniem z którym AI może sobie poradzić nawet lepiej niż inteligencja człowieka, ponieważ jest to dość skomplikowane do zoptymalizowania.
* Aby pisać proste programy wystarczy sztuczna inteligencja, ale do bardziej skomplikowanych zadań będzie wymagana inteligencja ludzka.
* Programowanie kanałów telewizyjnych nie wymaga inteligencji ludzkiej.
* Inteligencja ludzka jest konieczna do testowania skomplikowanych programów komputerowych, ludzie radzą sobie o wiele lepiej z odnajdywaniem błędów.
* Komponowanie muzyki nie wymaga ludzkiej inteligencji, ale moim zdaniem czynnik ludzki jest i tak niezbędny aby stworzyć coś z rzeczywistą duszą.
* Uważam, że wszystkie matematyczne działania mogą być wykonane przez AI. Rozwiązywanie wszystkich z tych działań opiera się na pracy na algorytmach oraz danych, co pozwala sztucznej inteligencji bez problemu je rozwiązywać.
* Kierowanie pojazdem już jest wykonywane przez AI, aczkolwiek moim zdaniem inteligencja ludzka (przynajmniej u niektórych) pozwala na lepsze przewidywanie potencjalnych sytuacji na drodze. Wydaje mi się, że w przyszłości jednak AI będzie prowadziło auta na tyle dobrze jak człowiek.

 Moim zdaniem główną różnicą pomiędzy botami, które mają za zadanie udawać człowieka, a tymi, które zaprogramowano na "asystentów", jest taka, że te drugie przez swoje zakodowanie są o wiele bardziej ograniczone. W przypadku czatbota przygotowanego na test Turinga używałem Cleverbota i pomimo, że nie jest to najbardziej zaawansowany system, to radził sobie o niebo lepiej niż Messengerowy bot promujący przypadkową firmę.

3. ***Sporządź raport ze spostrzeżeń.***

Do moich testów użyłem Cleverbota i jednego z messengerowych botów promujących przypadkową firmę. O dziwo bardziej sensowną konwersację dało się prowadzić z tą drugą maszyną. Tak długo jak temat rozmowy utrzymywał się w ramach firmy, którą bot promował, cała konwersacja nie była zła. Cleverbot natomiast praktycznie od samego początku ponosi kompletne fiasko w udawaniu człowieka. Jego odpowiedzi są niezwykle nietrafione, powtarzają się, bardzo często przeczy samemu sobie. Krótka rozmowa z Cleverbotem pokazała mi jak długą drogę przebyło AI od tamtych czasów, ponieważ dziś z Chatem GPT da się niejako odbyć rozmowę na jako takim poziomie. Z kolei dyskusja z botem z messengera pomogła mi docenić jak zaprogramowane jest to narzędzie, ponieważ uważam, że jego odpowiedzi mogłyby okazać się pomocne dla kogoś kto poszukuje informacji o danej firmie.
